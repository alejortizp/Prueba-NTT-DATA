```python
import numpy as np
import pandas as pd
```

```python
df_coffe = pd.read_parquet('coffee_db.parquet', engine='pyarrow')
df_coffe.shape
```

```python
df_coffe.columns
```

```python
df_coffe.head()
```

```python
df_coffe['Country'].nunique()
```

```python
df_coffe['Coffee type'].nunique()
```

```python
var_anios = ['1994/95', '1995/96', '1996/97', '1997/98', '1998/99', '1999/00',
       '2000/01', '2001/02', '2002/03', '2003/04', '2004/05', '2005/06',
       '2006/07', '2007/08', '2008/09', '2009/10', '2010/11', '2011/12',
       '2012/13', '2013/14', '2014/15', '2015/16', '2016/17', '2017/18',
       '2018/19', '2019/20']
```

```python
df_coffe.set_index(['Country'], inplace=True) #'Coffee type'
```

```python
df_coffe.head()
```

```python
df_coffe[var_anios].T.head() #[df_coffe['Country'] == 'Angola']
```

```python
df_t_coffe = df_coffe[var_anios].T
df_t_coffe.shape
```

```python
df_t_coffe.head()
```

```python
df_t_coffe.index
```

```python
cols = (df_t_coffe.columns)
print(cols)
```

```python
df_t_coffe['Fecha'] = pd.to_datetime(df_t_coffe.index, format='%Y/%y')
```

```python
df_t_coffe[['Angola', 'Malawi', 'Fecha']].head()
```

```python
# Usar melt para transformar el DataFrame
df_melted = df_t_coffe.melt(id_vars=["Fecha"], value_vars=cols, var_name="Country", value_name="Ventas")
df_melted.shape
```

```python
df_melted.head()
```

```python
df_coffe.reset_index(inplace=True)
```

```python
df_final = pd.merge(df_melted, df_coffe[['Country', 'Coffee type']], on='Country', how='left')
df_final.shape
```

```python
df_final.head()
```

```python
df_final.to_csv('coffee_final.csv', index=False)
```
